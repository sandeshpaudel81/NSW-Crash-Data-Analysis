Natural lighting values marked 'Unknown' were imputed based on two-hour interval mappings derived from general daylight timing.

```python
natural_lighting_imputation = {
    '16:00 - 17:59':'Dusk',
    '14:00 - 15:59':'Daylight',
    '12:00 - 13:59':'Daylight',  
    '10:00 - 11:59':'Daylight',
    '08:00 - 09:59':'Daylight',   
    '18:00 - 19:59':'Darkness',
    '06:00 - 07:59':'Dawn',
    '20:00 - 21:59':'Darkness',
    '22:00 - 00:00':'Darkness',
    '04:00 - 05:59':'Darkness',
    '00:01 - 01:59':'Darkness',
    '02:00 - 03:59':'Darkness'
}

def impute_natural_lighting(row):
    if row.natural_lighting == 'Unknown' and row.two_hour_intervals in natural_lighting_imputation:
        row.natural_lighting = natural_lighting_imputation[row.two_hour_intervals]
    return row

df_selected = df_selected.apply(impute_natural_lighting, axis=1)
```

```python
df_selected.natural_lighting.value_counts()
```

```python
df_selected[df_selected.two_hour_intervals == 'Unknown']
```

```python
two_hour_intervals_imputation = {
    'Dusk':'16:00 - 17:59',
    'Daylight': '14:00 - 15:59',
    'Darkness':'18:00 - 19:59',
    'Dawn': '06:00 - 07:59'
}

def impute_two_hour_intervals(row):
    if row.two_hour_intervals == 'Unknown' and row.natural_lighting in two_hour_intervals_imputation:
        row.two_hour_intervals = two_hour_intervals_imputation[row.natural_lighting]
    return row

df_selected = df_selected.apply(impute_two_hour_intervals, axis=1)
```
